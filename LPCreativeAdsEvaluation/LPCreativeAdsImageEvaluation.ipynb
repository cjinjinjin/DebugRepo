```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from collections import Counter
import numpy as np

# --- 辅助函数：类型转换和多数投票 ---

# def to_bool_or_na(series):
#     """健壮地将各种形式的值转换为 True, False, 或 pd.NA。"""
#     series_str = series.astype(str).str.lower().str.strip()
#     is_true = series_str.isin(['true', '1', 't'])
#     is_false = series_str.isin(['false', '0', 'f'])
#     return np.where(is_true, True, np.where(is_false, False, pd.NA))
def to_bool_or_na(series):
    """
    健壮地将各种形式的值转换为 True, False, 或 pd.NA。
    根据您的业务规则处理复杂的 HasHuman 值。
    """
    series_str = series.astype(str).str.lower().str.strip()
    
    # --- 1. 定义 TRUE 集合 (所有包含人的情况) ---
    # 包括：True 标准值、True 噪音格式、Partial、Silhouette
    true_patterns = [
        # 标准 True
        'true', 't', '1', 
        # 复杂 True 格式
        'true##!true', 'true (silhouette)', 
        # 部分人像 (Partial)
        'partial (hand only)', 'partial (hands only)'
    ]
    is_true = series_str.isin(true_patterns)
    
    # --- 2. 定义 FALSE 集合 (所有明确不包含人的情况) ---
    # 包括：False 标准值、False 噪音格式
    false_patterns = [
        # 标准 False
        'false', 'f', '0', 
        # 复杂 False 格式
        'false##!false', 'false>'
    ]
    is_false = series_str.isin(false_patterns)
    
    # --- 3. 返回结果 ---
    # 如果属于 True 集合，返回 True
    # 否则，如果属于 False 集合，返回 False
    # 否则 (例如 '空白'、None、或其他未识别的噪音)，返回 pd.NA
    return np.where(is_true, True, np.where(is_false, False, pd.NA))

def majority_vote_agg(series):
    """
    实现多数投票逻辑：如果 True 的票数 > False 的票数，返回 True。否则，返回 False。
    (忽略 NA/空值)
    """
    valid_votes = series.dropna()
    if valid_votes.empty:
        # 如果所有值都是 NA，我们返回 False 作为一个默认的安全值
        return False 
        
    true_count = valid_votes.sum()
    false_count = len(valid_votes) - true_count
    
    # 多数投票：True 票数严格大于 False 票数
    return true_count > false_count


def analyze_merged_metadata(tsv_filepath, image_folder_path, url_hash_col='UrlHash', tag_col='Tag', has_human_col='HasHuman', is_service_col='IsService', is_product_col='IsProduct', ad_category_col='L2'):
    """
    读取TSV文件，对 IsService/IsProduct 进行 UrlHash 级别多数投票整合，
    然后将整合结果合并回原始 Tag 级别数据，并统计组合分布。
    """
    
    # --- 1. 读取 TSV 文件并准备数据 ---
    try:
        df_raw = pd.read_csv(tsv_filepath, sep='\t')
    except FileNotFoundError:
        print(f"错误：未找到文件 {tsv_filepath}")
        return {}
    
    required_cols = [url_hash_col, tag_col, has_human_col, is_service_col, is_product_col]
    if not all(col in df_raw.columns for col in required_cols):
        missing = [col for col in required_cols if col not in df_raw.columns]
        print(f"错误：TSV 文件中缺少必需的列：{missing}")
        return {}

    # --- 2. 数据清洗和类型转换 ---
    # 对所有目标列进行布尔值转换 (HasHuman是Tag级的，Service/Product用于UrlHash级整合)
    for col in [has_human_col, is_service_col, is_product_col]:
        df_raw[col] = to_bool_or_na(df_raw[col])


    # --- 3. UrlHash 级别整合 (多数投票) ---
    consolidation_cols = [is_service_col, is_product_col]
    
    # 按 UrlHash 分组，对 IsService 和 IsProduct 进行多数投票
    df_consolidated = df_raw.groupby(url_hash_col)[consolidation_cols].agg(majority_vote_agg)
    
    # 重命名整合后的列，避免与原始列名冲突
    df_consolidated = df_consolidated.rename(columns={
        is_service_col: f'{is_service_col}_Consolidated',
        is_product_col: f'{is_product_col}_Consolidated'
    })
    
    print(f"UrlHash 级别整合完成。唯一 UrlHash 数: {len(df_consolidated)}")


    # --- 4. 合并数据：将整合后的特征填充回原始 Tag 级别数据 ---
    # 使用左连接 (left merge) 将整合结果匹配回原始数据 df_raw
    df_merged = df_raw.merge(df_consolidated, on=url_hash_col, how='left')


    # --- 5. 扫描图片文件夹，确定哪些 UrlHash_Tag 组合存在图片 ---
    
    # 创建一个 UrlHash_Tag 组合的集合，用于精确筛选
    image_key_set = set()
    try:
        for filename in os.listdir(image_folder_path):
            if filename.endswith(('.png', '.jpg', '.jpeg', '.webp')):
                # 假设图片名是 UrlHash_Tag.png
                parts = os.path.splitext(filename)[0].split('_', 1)
                if len(parts) == 2:
                    url_hash, tag = parts
                    image_key_set.add((url_hash, tag))
    except FileNotFoundError:
        print(f"错误：未找到图片文件夹 {image_folder_path}")
        return {}
    
    print(f"成功从图片文件夹中提取了 {len(image_key_set)} 个唯一的 UrlHash_Tag 组合。")
    
    
    # --- 6. 精确筛选和统计 ---
    
    # 创建一个用于筛选的 key 列
    df_merged['Image_Key'] = list(zip(df_merged[url_hash_col], df_merged[tag_col]))
    
    # 筛选出实际有图片文件的行
    df_filtered = df_merged[df_merged['Image_Key'].isin(image_key_set)].copy()
    
    if df_filtered.empty:
        print("警告：没有找到 TSV 数据与图片文件匹配的行。")
        return {}
    
    # 最终用于统计的列组合
    final_cols = [
        has_human_col, 
        f'{is_service_col}_Consolidated', 
        f'{is_product_col}_Consolidated'
    ]
    
    # 最终统计：基于 (HasHuman_Tag, IsService_UrlHash, IsProduct_UrlHash) 组合
    combination_counts = df_filtered.groupby(final_cols).size()
    category_counts = df_filtered[ad_category_col].value_counts().to_dict()

    # 将结果转换为字典
    return dict(combination_counts.to_dict()), category_counts

# --- 示例用法 (请替换为您的真实路径) ---

TSV_FILE = r"D:\Data\T2I\AutoGenValidation\2025-11-23-Validation.GPT5Prompts-4kToken-expand.tsv"
IMAGE_DIR = r"D:\Data\T2I\AutoGenValidation\recraft_hasDefect"

if __name__ == '__main__':
    # ----------------------------------------------------

    results, category_distribution = analyze_merged_metadata(TSV_FILE, IMAGE_DIR, ad_category_col='L1')

    if results:
        print("\n--- 最终统计结果 (Tag 级 HasHuman + UrlHash 级 Service/Product) ---")
        print("格式: (HasHuman_Tag, IsService_UrlHash, IsProduct_UrlHash): 计数")
        
        # 结果键名：(HasHuman, IsService_Consolidated, IsProduct_Consolidated)
        # ----------------------------------------------------------------------------------
        # h1_t1: HasHuman=True, IsService=False, IsProduct=False (UrlHash h1 整合结果) -> 1 次
        # h3_t4: HasHuman=True, IsService=True, IsProduct=False (UrlHash h3 整合结果) -> 1 次
        # h3_t5: HasHuman=False, IsService=True, IsProduct=False (UrlHash h3 整合结果) -> 1 次
        # ----------------------------------------------------------------------------------
        
        for combo, count in sorted(results.items(), key=lambda item: item[1], reverse=True):
            print(f"{combo}: {count}")

        print("\n--- L2 广告类别标签分布 (仅针对有图片文件的行) ---")
        print(f"总计行数: {sum(category_distribution.values())}")
        for category, count in sorted(category_distribution.items(), key=lambda item: item[1], reverse=True):
            print(f"'{category}': {count}")
        
                
        # ------------------------------------------------------------------
        # 1. 使用您上面打印的 df_category DataFrame
        # ------------------------------------------------------------------
        total_count = sum(category_distribution.values())
        
        # 将计数字典转换为DataFrame，方便计算百分比和排序
        df_category = pd.DataFrame(
            list(category_distribution.items()), 
            columns=['Category', 'Count']
        )
        df_category['Percentage'] = (df_category['Count'] / total_count) * 100
        
        # 按百分比降序排序
        df_category = df_category.sort_values(by='Percentage', ascending=False)
        df_plot = df_category # 直接使用前面计算的 DataFrame

        # 设置画布大小
        plt.figure(figsize=(12, 6))

        # 创建柱状图，使用 Percentage 作为高度
        bars = plt.bar(df_plot['Category'], df_plot['Percentage'], color='skyblue')

        # 添加标签、标题和百分比格式化
        plt.title('L1 Distribution of defect images(Recraft)', fontsize=16)
        plt.xlabel('L1', fontsize=12)
        plt.ylabel('Percentage (%)', fontsize=12)
        plt.xticks(rotation=45, ha='right') # 旋转X轴标签以防重叠

        # 在每个柱子上方添加百分比文本标签
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, 
                    f'{yval:.1f}%', ha='center', va='bottom')

        plt.tight_layout() # 调整布局，防止标签溢出
        plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np

# --- 辅助函数：类型转换和多数投票 (保持不变) ---
# (to_bool_or_na 和 majority_vote_agg 函数定义保持不变)

def to_bool_or_na(series):
    """
    健壮地将各种形式的值转换为 True, False, 或 pd.NA。
    根据您的业务规则处理复杂的 HasHuman 值。
    """
    series_str = series.astype(str).str.lower().str.strip()
    
    # --- 1. 定义 TRUE 集合 (所有包含人的情况) ---
    true_patterns = [
        'true', 't', '1', 
        'true##!true', 'true (silhouette)', 
        'partial (hand only)', 'partial (hands only)'
    ]
    is_true = series_str.isin(true_patterns)
    
    # --- 2. 定义 FALSE 集合 (所有明确不包含人的情况) ---
    false_patterns = [
        'false', 'f', '0', 
        'false##!false', 'false>'
    ]
    is_false = series_str.isin(false_patterns)
    
    # --- 3. 返回结果 ---
    return np.where(is_true, True, np.where(is_false, False, pd.NA))

def majority_vote_agg(series):
    """
    实现多数投票逻辑：如果 True 的票数 > False 的票数，返回 True。否则，返回 False。
    (忽略 NA/空值)
    """
    valid_votes = series.dropna()
    if valid_votes.empty:
        return False 
        
    true_count = valid_votes.sum()
    false_count = len(valid_votes) - true_count
    
    return true_count > false_count


def analyze_merged_metadata(tsv_filepath, image_folder_path, url_hash_col='UrlHash', tag_col='Tag', has_human_col='HasHuman', is_service_col='IsService', is_product_col='IsProduct', ad_category_col='L2'):
    """
    读取TSV文件，对 IsService/IsProduct 进行 UrlHash 级别多数投票整合，
    并计算 L1/L2 广告类别在所有 TSV 原始行中的绝对覆盖率。
    """
    
    # --- 1. 读取 TSV 文件并准备数据 ---
    try:
        df_raw = pd.read_csv(tsv_filepath, sep='\t')
    except FileNotFoundError:
        print(f"错误：未找到文件 {tsv_filepath}")
        return {}, {}
    
    # 检查必需列，包括 ad_category_col
    required_cols = [url_hash_col, tag_col, has_human_col, is_service_col, is_product_col, ad_category_col]
    if not all(col in df_raw.columns for col in required_cols):
        missing = [col for col in required_cols if col not in df_raw.columns]
        print(f"错误：TSV 文件中缺少必需的列：{missing}")
        return {}, {}

    # --- 2. 数据清洗和类型转换 ---
    for col in [has_human_col, is_service_col, is_product_col]:
        df_raw[col] = to_bool_or_na(df_raw[col])
    
    # Ad Category (L1/L2) 列：将空值统一转换为字符串 'UNKNOWN_CATEGORY'
    df_raw[ad_category_col] = df_raw[ad_category_col].fillna('UNKNOWN_CATEGORY').astype(str).str.strip()
    
    # --- 2.5 关键：计算 TSV 原始总行数 (UrlHash + Tag 组合总数) 作为分母 ---
    total_raw_rows = len(df_raw)
    print(f"TSV 原始总行数 (分母): {total_raw_rows}")

    # --- 3. UrlHash 级别整合 & 4. 合并数据 (保持不变) ---
    consolidation_cols = [is_service_col, is_product_col]
    df_consolidated = df_raw.groupby(url_hash_col)[consolidation_cols].agg(majority_vote_agg)
    service_col_c = f'{is_service_col}_Consolidated'
    product_col_c = f'{is_product_col}_Consolidated'
    df_consolidated = df_consolidated.rename(columns={is_service_col: service_col_c, is_product_col: product_col_c})
    df_merged = df_raw.merge(df_consolidated, on=url_hash_col, how='left')


    # --- 5. 扫描图片文件夹 & 6. 精确筛选 (保持不变) ---
    image_key_set = set()
    try:
        for filename in os.listdir(image_folder_path):
            if filename.endswith(('.png', '.jpg', '.jpeg', '.webp')):
                parts = os.path.splitext(filename)[0].split('_', 1)
                if len(parts) == 2:
                    image_key_set.add((parts[0], parts[1]))
    except FileNotFoundError:
        print(f"错误：未找到图片文件夹 {image_folder_path}")
        return {}, {}
    
    print(f"成功从图片文件夹中提取了 {len(image_key_set)} 个唯一的 UrlHash_Tag 组合 (分子总和): {len(image_key_set)}")
    
    df_merged['Image_Key'] = list(zip(df_merged[url_hash_col], df_merged[tag_col]))
    df_filtered = df_merged[df_merged['Image_Key'].isin(image_key_set)].copy()
    
    if df_filtered.empty:
        print("警告：没有找到 TSV 数据与图片文件匹配的行。")
        return {}, {}
    
    # 最终统计：组合分布 (保持不变)
    final_cols = [has_human_col, service_col_c, product_col_c]
    df_filtered[has_human_col] = df_filtered[has_human_col].fillna(False)
    combination_counts = df_filtered.groupby(final_cols).size().to_dict()
    
    # --- 7. L1/L2 绝对覆盖率计算 (新的计算逻辑) ---
    
    # 计算有图片文件中的 Category 计数 (分子)
    filtered_category_counts = df_filtered[ad_category_col].value_counts().to_dict()

    # 创建最终的 Category 统计 DataFrame
    final_category_df = pd.DataFrame(
        {'Category': filtered_category_counts.keys(), 'Filtered_Count': filtered_category_counts.values()}
    )
    
    # 计算百分比：(Filtered_Count / total_raw_rows) * 100
    final_category_df['Coverage_Percentage'] = (final_category_df['Filtered_Count'] / total_raw_rows) * 100
    
    # 转换为字典格式返回
    category_results = final_category_df.set_index('Category')[['Filtered_Count', 'Coverage_Percentage']].to_dict(orient='index')
    
    return combination_counts, category_results

# --- 示例用法 (保持不变) ---

TSV_FILE = r"D:\Data\T2I\AutoGenValidation\2025-11-23-Validation.GPT5Prompts-4kToken-expand.tsv"
IMAGE_DIR = r"D:\Data\T2I\AutoGenValidation\recraft_hasDefect"

if __name__ == '__main__':
    CATEGORY_COL = 'L1' 
    
    
    combination_results, category_results = analyze_merged_metadata(TSV_FILE, IMAGE_DIR, ad_category_col=CATEGORY_COL)

    if results and category_results:
        # 1. 组合分布结果打印 (省略)
        print("\n--- 最终统计结果 (HasHuman/Service/Product 组合) ---")
        for combo, count in sorted(results.items(), key=lambda item: item[1], reverse=True):
            print(f"{combo}: {count}")
            
        # 2. L1 绝对覆盖率绘图
        print("\n--- L1 广告类别绝对覆盖率 (分母为 TSV 原始总行数) ---")
        
        # 将返回的字典转换为 DataFrame
        df_category_coverage = pd.DataFrame.from_dict(category_results, orient='index').reset_index().rename(columns={'index': 'Category'})
        df_category_coverage = df_category_coverage.sort_values(by='Coverage_Percentage', ascending=False)
        
        # 预期百分比 (总分母=7): Tech: 2/7=28.57%, Finance: 1/7=14.29%, E-com: 1/7=14.29%
        # print(f"\n[用于绘图的标准化数据 - 绝对覆盖率，TSV 原始总行数: {len(mock_df)}]")
        # print(df_category_coverage.to_markdown(index=False, floatfmt=".2f"))

        # ------------------------------------------------------------------
        # 绘图部分：L1 绝对覆盖率百分比柱状图
        # ------------------------------------------------------------------
        
        df_plot = df_category_coverage # 使用包含绝对覆盖率的数据

        # 设置画布大小
        plt.figure(figsize=(12, 6))
        
        # 创建柱状图，使用 Coverage_Percentage 作为高度
        bars = plt.bar(df_plot['Category'], df_plot['Coverage_Percentage'], color='indianred')
        
        # 添加标签、标题和百分比格式化
        plt.title(f'{CATEGORY_COL} defect rate of Recraft', fontsize=16)
        plt.xlabel(CATEGORY_COL, fontsize=12)
        plt.ylabel('Absolute Percentage (%)', fontsize=12)
        plt.xticks(rotation=45, ha='right') # 旋转X轴标签以防重叠
        
        # 在每个柱子上方添加百分比文本标签
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, 
                     f'{yval:.1f}%', ha='center', va='bottom')
        
        plt.tight_layout() # 调整布局，防止标签溢出
        plt.show() # 显示图表
```

```python
import os
import csv

def match_images_with_prompts(image_dir, prompt_file, output_file):
    # 1. 获取文件夹下所有 .png 文件名（不含后缀，方便匹配）
    image_files = {f for f in os.listdir(image_dir) if f.lower().endswith('.png')}
    print(f"找到 {len(image_files)} 张 PNG 图片用于匹配。", image_files)
    results = []
    
    # 2. 读取 Prompt 文件
    # 这里假设你的 prompt_file 是以逗号或制表符分隔的，例如：filename,prompt
    with open(prompt_file, 'r', encoding='utf-8') as f:
        for line in f:
            # 根据你的文件格式调整分隔符，这里以逗号为例
            parts = line.strip().split('\t', 1) 
            if len(parts) < 2:
                continue
                
            name_in_file, prompt = parts[0].strip(), parts[1].strip()
            # 自动补齐后缀进行匹配（或者去掉后缀匹配）
            if f"{name_in_file}.png" in image_files:
                results.append((name_in_file, prompt))
            
    print(f"匹配到 {len(results)} 条结果。")
    # 3. 输出结果
    with open(output_file, 'w', encoding='utf-8') as out:
        for filename, prompt in results:
            out.write(f"{filename}.png\t{filename}_NegativePrompt1.png\t{prompt}\n")
            out.write(f"{filename}.png\t{filename}_NegativePrompt2.png\t{prompt}\n")
            out.write(f"{filename}.png\t{filename}_NegativePrompt3.png\t{prompt}\n")
            out.write(f"{filename}.png\t{filename}_NegativePrompt4.png\t{prompt}\n")
            out.write(f"{filename}.png\t{filename}_NegativePrompt5.png\t{prompt}\n")
            print(f"已匹配: {filename}")

# --- 配置区域 ---
IMAGE_FOLDER = r"D:\Data\T2I\NanoBanana\super-realism-prompts-NanoBanana\score_le_0.5"      # 你的图片文件夹路径
PROMPT_LIST_FILE = r"D:\Data\T2I\ZImage\official\super-realism-prompts1k_official_20260116-0314\analysis_results\super-realism-prompts-new_withId.tsv" # 你的 Prompt 列表文件
OUTPUT_RESULT = r"D:\Data\T2I\ZImage\official\super-realism-prompts1k_official_20260116-0314\analysis_results\score_le_0.5_nanobanana_result.txt"  # 输出结果文件

if __name__ == "__main__":
    match_images_with_prompts(IMAGE_FOLDER, PROMPT_LIST_FILE, OUTPUT_RESULT)
```

```python
import pandas as pd

# 1. 加载数据
df = pd.read_csv(r"C:\Users\jinjinchen\Downloads\UHRS_Task_lp_relevance_labeling_0122.csv")

def dpo_vote_logic(series):
    # 计算 Bad 出现的次数
    bad_count = (series == 'Bad').sum()
    # 只有 2 票或 3 票 Bad 时，最终共识才为 Bad
    if bad_count >= 2:
        return 'Bad'
    # 其他情况（包括 1:1:1 或 Good/Fair 占优）统一记为 Good
    return 'Good'

# 应用新的投票逻辑
consensus = df.groupby(['HitID', 'ImageSource'])['Rating'].agg(dpo_vote_logic).reset_index()

# 3. 统计每个 Source 的具体数量和比例
def get_detailed_stats(group):
    total = len(group)
    bad_num = (group == 'Bad').sum()
    good_num = (group == 'Good').sum()
    bad_rate = (bad_num / total) * 100
    
    # 返回一个 Series，方便最后直接生成表格
    return pd.Series({
        'Total': total,
        'Bad_Count': bad_num,
        'Good_Count': good_num,
        'Bad_Rate(%)': round(bad_rate, 2)
    })

# 分组应用统计函数
result = consensus.groupby('ImageSource')['Rating'].apply(get_detailed_stats).unstack()

print("--- 各数据源 标注统计结果 ---")
print(result)

# 如果你需要导出这个统计表到 PPT
# result.to_csv("dpo_stat_summary.csv")
```

```python
import pandas as pd
import re

# 1. 读取数据（假设你已另存为 CSV）
df = pd.read_csv(r"C:\Users\jinjinchen\Downloads\UHRS_Task_lp_relevance_labeling_0122.csv", header=0)
df.columns = df.columns.str.strip()

# 2. 投票函数：3取2，平票(1:1:1)记为Bad
def get_vote(series):
    counts = series.value_counts()
    if counts.iloc[0] >= 2:
        return counts.index[0]
    return 'Bad'

# 3. 第一步聚合：先计算每条数据的投票共识结果
# 必须保留 ImgUrl，因为我们要从这里提取 ID
consensus_df = df.groupby(['HitID', 'ImgUrl', 'ImageSource'])['Rating'].apply(get_vote).reset_index()

# 4. 提取 ImageID 和 Prompt 编号
# 针对格式：.../adst2iv3_52_Prompt5.jpg
def extract_info(url):
    # 提取 ID 部分（adst2iv3_52）
    id_match = re.search(r'([^/]+)_Prompt\d', url)
    # 提取 Prompt 编号（Prompt5）
    prompt_match = re.search(r'(Prompt\d)', url)
    
    image_id = id_match.group(1) if id_match else "Unknown"
    prompt_label = prompt_match.group(1) if prompt_match else "Unknown"
    return pd.Series([image_id, prompt_label])

consensus_df[['ImageID', 'Prompt']] = consensus_df['ImgUrl'].apply(extract_info)

# 5. 过滤出 AdsT2Iv3 的数据进行 ID level 分析
ads_df = consensus_df[consensus_df['ImageSource'] == 'AdsT2Iv3'].copy()

# 6. ID Level 统计：检查每个 ImageID 下的 Prompt 表现
def check_all_bad(group):
    # 检查是否 5 个 Prompt 都到齐了（可选，根据数据完整性决定）
    # 检查是否该 ID 下所有的 Rating 都是 Bad
    is_all_bad = (group == 'Bad').all()
    return is_all_bad

id_level_results = ads_df.groupby('ImageID')['Rating'].apply(check_all_bad).reset_index()
id_level_results.rename(columns={'Rating': 'Is_All_Bad'}, inplace=True)

# 7. 计算最终比例
total_ids = len(id_level_results)
all_bad_ids = id_level_results['Is_All_Bad'].sum()
all_bad_rate = (all_bad_ids / total_ids * 100) if total_ids > 0 else 0

print(f"--- AdsT2Iv3 Source ID Level 分析 ---")
print(f"总 Image ID 数量: {total_ids}")
print(f"Prompt 1-5 均为 Bad 的 ID 数量: {all_bad_ids}")
print(f"全 Bad 比例: {all_bad_rate:.2f}%")

all_bad_ids_list = id_level_results[id_level_results['Is_All_Bad'] == True]['ImageID']

# 2. 从原始 consensus_df 中找到这些 ID 对应的所有行
# 这样你可以看到每个 ID 对应的 5 个 Prompt URL
all_bad_urls_df = ads_df[ads_df['ImageID'].isin(all_bad_ids_list)]

# 3. 排序，让同一个 ID 的 Prompt 1-5 排在一起，方便观察
all_bad_urls_df = all_bad_urls_df.sort_values(by=['ImageID', 'Prompt'])

# 4. 输出到控制台或保存为 Excel/CSV
print(f"找到 {len(all_bad_ids_list)} 个 ID 的所有 Prompt 均为 Bad，共计 {len(all_bad_urls_df)} 条链接。")

# 如果你想直接看 URL 列表
print("\n部分全 Bad 图片链接示例：")
print(all_bad_urls_df[['ImageID', 'Prompt', 'ImgUrl']])
```

```python
import pandas as pd
import re
import os
import requests  # 需要安装: pip install requests
from urllib.parse import urlparse

# 1. 读取数据
df = pd.read_csv(r"C:\Users\jinjinchen\Downloads\UHRS_Task_lp_relevance_labeling_0122.csv", header=0)
df.columns = df.columns.str.strip()

# 2. 投票函数
def get_vote(series):
    counts = series.value_counts()
    if counts.iloc[0] >= 2:
        return counts.index[0]
    return 'Bad'

# 3. 第一步聚合
consensus_df = df.groupby(['HitID', 'ImgUrl', 'ImageSource'])['Rating'].apply(get_vote).reset_index()

# 4. 提取信息
def extract_info(url):
    id_match = re.search(r'([^/]+)_Prompt\d', url)
    prompt_match = re.search(r'(Prompt\d)', url)
    image_id = id_match.group(1) if id_match else "Unknown"
    prompt_label = prompt_match.group(1) if prompt_match else "Unknown"
    return pd.Series([image_id, prompt_label])

consensus_df[['ImageID', 'Prompt']] = consensus_df['ImgUrl'].apply(extract_info)

# 5. 过滤数据
ads_df = consensus_df[consensus_df['ImageSource'] == 'AdsT2Iv3'].copy()

# 6. ID Level 统计：找到所有 Prompt 均为 Bad 的 ID
def check_all_bad(group):
    return (group == 'Bad').all()

id_level_results = ads_df.groupby('ImageID')['Rating'].apply(check_all_bad).reset_index()
id_level_results.rename(columns={'Rating': 'Is_All_Bad'}, inplace=True)

# 7. 筛选出需要下载的 URL 列表
all_bad_ids_list = id_level_results[id_level_results['Is_All_Bad'] == True]['ImageID']
all_bad_urls_df = ads_df[ads_df['ImageID'].isin(all_bad_ids_list)].sort_values(by=['ImageID', 'Prompt'])

print(f"找到 {len(all_bad_ids_list)} 个全 Bad ID，共计 {len(all_bad_urls_df)} 张图片准备下载。")

# --- 新增下载部分 ---

# 8. 配置下载目录
download_dir = r"C:\Users\jinjinchen\Downloads\AdsT2I_Bad_Images"
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

def download_images(dataframe, save_path):
    print(f"开始下载图片到: {save_path} ...")
    success_count = 0
    fail_count = 0
    
    for index, row in dataframe.iterrows():
        url = row['ImgUrl']
        image_id = row['ImageID']
        prompt = row['Prompt']
        
        # 构造文件名: ImageID_Prompt.jpg
        file_extension = os.path.splitext(urlparse(url).path)[1]
        if not file_extension: file_extension = ".jpg" # 缺省后缀
        filename = f"{image_id}_{prompt}{file_extension}"
        full_path = os.path.join(save_path, filename)
        
        try:
            # 设置超时以防挂死
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                with open(full_path, 'wb') as f:
                    f.write(response.content)
                success_count += 1
                if success_count % 10 == 0:
                    print(f"已成功下载 {success_count} 张...")
            else:
                print(f"下载失败 (HTTP {response.status_code}): {url}")
                fail_count += 1
        except Exception as e:
            print(f"下载异常: {url}, 错误: {e}")
            fail_count += 1

    print(f"\n下载完成！")
    print(f"成功: {success_count}")
    print(f"失败: {fail_count}")

# 执行下载
if not all_bad_urls_df.empty:
    download_images(all_bad_urls_df, download_dir)
else:
    print("没有发现符合条件的全 Bad 图片。")
```

```python
import pandas as pd
import os
import requests
from concurrent.futures import ThreadPoolExecutor

def vote_logic(series):
    # 计算 Bad 出现的次数
    bad_count = (series == 'Bad').sum()
    # 只有 2 票或 3 票 Bad 时，最终共识才为 Bad
    if bad_count >= 2:
        return 'Bad'
    # 其他情况（包括 1:1:1 或 Good/Fair 占优）统一记为 Good
    return 'Good'

# 1. 加载数据
df = pd.read_csv(r"C:\Users\jinjinchen\Downloads\UHRS_Task_lp_relevance_labeling_0122.csv")
df.columns = df.columns.str.strip()

# 2. 严格对齐之前的投票逻辑：以 HitID 为准取众数
# 这一步是为了找回那 84 个样本
# consensus = df.groupby(['HitID', 'ImgUrl', 'ImageSource'])['Rating'].agg(lambda x: x.mode()[0]).reset_index()
consensus = df.groupby(['HitID', 'ImgUrl', 'ImageSource'])['Rating'].agg(vote_logic).reset_index()

# 3. 筛选出 AdsT2Iv3 中投票结果为 Bad 的数据
# 这里应该正好就是 84 条记录
bad_samples_df = consensus[(consensus['ImageSource'] == 'AdsT2Iv3') & (consensus['Rating'] == 'Bad')].copy()

print(f"--- 统计核对 ---")
print(f"AdsT2Iv3 投票共识为 Bad 的样本数: {len(bad_samples_df)}") 

# 4. 下载配置
save_dir = r"C:\Users\jinjinchen\Downloads\AdsT2Iv3_Consensus_Bad_0127"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

def download_engine(row):
    # print(row)
    url = row['ImgUrl']
    # 提取文件名，处理可能的特殊字符
    file_name = url.split('/')[-1].split('?')[0] 
    save_path = os.path.join(save_dir, file_name)
    
    try:
        # 加上 Headers 模拟浏览器，防止某些 CDN 拦截
        headers = {'User-Agent': 'Mozilla/5.0'}
        r = requests.get(url, headers=headers, timeout=10)
        if r.status_code == 200:
            with open(save_path, 'wb') as f:
                f.write(r.content)
            return True
        else:
            return False
    except:
        return False

# 5. 执行多线程下载
if len(bad_samples_df) > 0:
    print(f"开始并行下载 {len(bad_samples_df)} 张图片...")
    with ThreadPoolExecutor(max_workers=10) as executor:
        results = list(executor.map(download_engine, [row for _, row in bad_samples_df.iterrows()]))
    print(f"下载完成！成功: {sum(results)} / 失败: {len(results) - sum(results)}")
else:
    print("未找到符合条件的 Bad 样本，请检查 ImageSource 名称或 Rating 标签内容。")
```

```python
import pandas as pd
import requests
import os

# ================= 配置区域 =================
input_file = r"D:\Data\T2I\AutoGenValidation\UHRS_Task_Ads_image_quality_labeling_0122.tsv"
base_save_dir = r"D:\Data\T2I\AutoGenValidation\UHRS_Task_Ads_image_quality_labeling_0122_bad_images"
target_source = 'AdsT2Iv3_LP'

priority_dimensions = [
    'TextLogoClarity', 
    'RealisticPhysicallyPlausible', 
    'BrightnessContrastColorSaturation', 
    'Imageclarity', 
    'SubjectClarity', 
    'CompositionLayout'
]

os.makedirs(base_save_dir, exist_ok=True)

# ================= 1. 数据清洗与聚合 =================
df = pd.read_csv(input_file, sep='\t', header=0)

# 只保留目标 Source 并清洗 URL 和 标签
df = df[df['ImageSource'] == target_source].copy()
df['ImgUrl'] = df['ImgUrl'].astype(str).str.strip()
df['FinalDecision'] = df['FinalDecision'].astype(str).str.strip()

# 数字化维度得分
for dim in priority_dimensions:
    df[dim] = df[dim].astype(str).str.strip()
    df[f"{dim}_score"] = (df[dim] == 'Bad').astype(int)

# 数字化总投票
df['Final_score'] = (df['FinalDecision'] == 'Bad').astype(int)

# ---【按 ImgUrl 聚合所有人的投票】---
agg_rules = {f"{dim}_score": 'sum' for dim in priority_dimensions}
agg_rules['Final_score'] = 'sum'

grouped = df.groupby('ImgUrl').agg(agg_rules).reset_index()

# ================= 2. 锁定 Bad 图片池 =================
# 只要该 URL 累计收到的 FinalDecision=Bad 票数 >= 2
bad_pool = grouped[grouped['Final_score'] >= 2].copy()

print(f"--- 最终统计 (仅基于 ImgUrl) ---")
print(f"聚合后唯一 URL 总数: {len(grouped)}")
print(f"最终判定为 Bad 的图片数: {len(bad_pool)}")

processed_urls = set()
analysis_summary = []

# ================= 3. 优先级不放回分拣 =================
for dim in priority_dimensions:
    dim_dir = os.path.join(base_save_dir, dim)
    os.makedirs(dim_dir, exist_ok=True)
    
    col_name = f"{dim}_score"
    
    # 筛选：Bad 图片池中 + 没被处理过 + 该维度共识 >= 2
    current_batch = bad_pool[
        (~bad_pool['ImgUrl'].isin(processed_urls)) & 
        (bad_pool[col_name] >= 2)
    ].copy()
    
    print(f"维度 [{dim}]: 分拣 {len(current_batch)} 张")
    
    for idx, row in current_batch.iterrows():
        url = row['ImgUrl']
        processed_urls.add(url)
        
        # 提取 URL 最后的名称作为文件名 (去掉 query 参数)
        pure_name = url.split('/')[-1].split('?')[0]
        
            
        save_path = os.path.join(dim_dir, pure_name)
        
        # [执行下载]
        # try:
        #     r = requests.get(url, timeout=10)
        #     if r.status_code == 200:
        #         with open(save_path, 'wb') as f:
        #             f.write(r.content)
        # except:
        #     pass
            
        analysis_summary.append({'ImgUrl': url, 'Category': dim})

# ================= 4. 处理原因不一致的 Bad 图片 =================
remaining_bad = bad_pool[~bad_pool['ImgUrl'].isin(processed_urls)].copy()

if len(remaining_bad) > 0:
    disputed_dir = os.path.join(base_save_dir, "Inconsistent_Reasons")
    os.makedirs(disputed_dir, exist_ok=True)
    print(f"原因不一致: 分拣 {len(remaining_bad)} 张")
    
    for idx, row in remaining_bad.iterrows():
        url = row['ImgUrl']
        processed_urls.add(url)
        pure_name = url.split('/')[-1].split('?')[0]
        save_path = os.path.join(disputed_dir, pure_name)
        
        # [执行下载]
        try:
            r = requests.get(url, timeout=10)
            if r.status_code == 200:
                with open(save_path, 'wb') as f:
                    f.write(r.content)
        except:
            pass
        
        analysis_summary.append({'ImgUrl': url, 'Category': 'Inconsistent_Reasons'})

# 保存最终映射关系
pd.DataFrame(analysis_summary).to_csv(os.path.join(base_save_dir, "url_to_category.csv"), index=False)
```

```python
import pandas as pd
import re

# 1. 读取数据（假设你已另存为 CSV）
df = pd.read_csv(r"D:\Data\T2I\AutoGenValidation\AdsT2I_0131-2\UHRS_Task_lp_relevance_labeling_0204_2.tsv", sep='\t', header=0)
df.columns = df.columns.str.strip()

# 2. 投票函数：3取2，平票(1:1:1)记为Bad
def get_vote(series):
    counts = series.value_counts()
    if counts.iloc[0] >= 2:
        return counts.index[0]
    return 'Bad'

# 3. 第一步聚合：先计算每条数据的投票共识结果
# 必须保留 ImgUrl，因为我们要从这里提取 ID
consensus_df = df.groupby(['HitID', 'ImgUrl', 'ImageSource'])['Rating'].apply(get_vote).reset_index()

# 4. 提取 ImageID 和 Prompt 编号
# 针对格式：.../adst2iv3_52_Prompt5.jpg
# https://unitorchazureblob.blob.core.windows.net/bic/Internal10066_Prompt4.png
def extract_info(url):
    # 提取 ID 部分（adst2iv3_52）
    id_match = re.search(r'([^/]+)_Prompt\d', url)
    # 提取 Prompt 编号（Prompt5）
    prompt_match = re.search(r'(Prompt\d)', url)
    
    image_id = id_match.group(1) if id_match else "Unknown"
    prompt_label = prompt_match.group(1) if prompt_match else "Unknown"
    print(f"提取信息 - URL: {url} | ImageID: {image_id} | Prompt: {prompt_label}")
    return pd.Series([image_id, prompt_label])

consensus_df[['ImageID', 'Prompt']] = consensus_df['ImgUrl'].apply(extract_info)

# 5. 过滤出 AdsT2Iv3 的数据进行 ID level 分析
ads_df = consensus_df[consensus_df['ImageSource'] == 'AdsT2Iv3'].copy()

# 6. ID Level 统计：检查每个 ImageID 下的 Prompt 表现
def check_all_bad(group):
    # 检查是否 5 个 Prompt 都到齐了（可选，根据数据完整性决定）
    # 检查是否该 ID 下所有的 Rating 都是 Bad
    is_all_bad = (group == 'Bad').all()
    return is_all_bad

id_level_results = ads_df.groupby('ImageID')['Rating'].apply(check_all_bad).reset_index()
id_level_results.rename(columns={'Rating': 'Is_All_Bad'}, inplace=True)

# 7. 计算最终比例
total_ids = len(id_level_results)
all_bad_ids = id_level_results['Is_All_Bad'].sum()
all_bad_rate = (all_bad_ids / total_ids * 100) if total_ids > 0 else 0

print(f"--- AdsT2Iv3 Source ID Level 分析 ---")
print(f"总 Image ID 数量: {total_ids}")
print(f"Prompt 1-5 均为 Bad 的 ID 数量: {all_bad_ids}")
print(f"全 Bad 比例: {all_bad_rate:.2f}%")

all_bad_ids_list = id_level_results[id_level_results['Is_All_Bad'] == True]['ImageID']

# 2. 从原始 consensus_df 中找到这些 ID 对应的所有行
# 这样你可以看到每个 ID 对应的 5 个 Prompt URL
all_bad_urls_df = ads_df[ads_df['ImageID'].isin(all_bad_ids_list)]

# 3. 排序，让同一个 ID 的 Prompt 1-5 排在一起，方便观察
all_bad_urls_df = all_bad_urls_df.sort_values(by=['ImageID', 'Prompt'])

# 4. 输出到控制台或保存为 Excel/CSV
print(f"找到 {len(all_bad_ids_list)} 个 ID 的所有 Prompt 均为 Bad，共计 {len(all_bad_urls_df)} 条链接。")

# 如果你想直接看 URL 列表
print("\n部分全 Bad 图片链接示例：")
print(all_bad_urls_df[['ImageID', 'Prompt', 'ImgUrl']])
```
